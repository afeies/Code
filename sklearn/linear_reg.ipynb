## Linear Regression

- import dataset
- import train test split

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

- use only one feature

```python
X, y = load_diabetes(return_X_y=True)
X = X[:, [2]]
X[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20, shuffle=False)
```

### linear regression model
- create model
- fit it on the training data

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression().fit(X_train, y_train)
```

### model evaluation
- evaluate on test set
    - mean squared error
    - coefficient of determination (r2_score)

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = regressor.predict(X_test)

print(f"Mean squared error: {mean_squared_error(y_test, y_pred):.2f}")
print(f"Coefficient of determination: {r2_score(y_test, y_pred):.2f}")
```

### Plotting the results
- on train and test data

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(10, 5), sharex=True, sharey=True)

# create scatter plot of points
ax[0].scatter(X_train, y_train, label="Train data points")

# plot the line
ax[0].plot(X_train, regressor.predict(X_train), color="tab:orange", linewidth=5, label="Model predictions")

# title and axis names
ax[0].set(xlabel="Feature", ylabel="Target", title="Train set")

# legend (dispays labels)
ax[0].legend()

ax[1].scatter(X_test, y_test, label="Test data points")
ax[1].plot(X_test, y_pred, color="tab:orange", linewidth=5, label="Model predictions")
ax[1].set(xlabel="Feature", ylabel="Target", title="Test set")
ax[1].legend()

fig.suptitle("Linear Regression")
```
